```python
umirovljenici4 = umirovljenici.copy()
```

```python
prosjek_godina = int(umirovljenici4["starost"].mean())
```

```python
prosjek_godina
```

```python
umirovljenici4["starost"] = umirovljenici4["starost"].fillna(prosjek_godina)
```

```python
umirovljenici4
```

```python
umirovljenici4["spol"].mode()[0]
```

```python
umirovljenici4["spol"] = umirovljenici4["spol"].fillna(value = "m")
```

```python
umirovljenici4["zanimanje"] = umirovljenici4["zanimanje"].fillna(value = umirovljenici4["zanimanje"].mode()[0])
```

```python
umirovljenici4
```

```python
umirovljenici4["primanja"] = umirovljenici4["primanja"].fillna(value = umirovljenici4["primanja"].median())
```

```python
umirovljenici4["stednja"] = umirovljenici4["stednja"].fillna(value = umirovljenici4["stednja"].median())
```
