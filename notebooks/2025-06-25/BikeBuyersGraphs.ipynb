```python
import pandas as pd
import matplotlib.pyplot as plt
bikes = pd.read_csv("bike_buyers_clean.csv")
```

```python
bikes.head()
```

```python
bikes.describe()
```

```python
bikes.corr(numeric_only=True)
```

```python
plt.xlabel('Occupation')
plt.ylabel('Income')
plt.bar(bikes.Occupation, bikes.Income)
```

```python
plt.xlabel('Broj djece')
plt.ylabel('Income')
plt.stem(bikes.Children, bikes.Income)
```

```python
plt.xlabel('Posjeduju bicikl')
plt.ylabel('Ne posjeduju bicikl')
purchased_bike=bikes["Purchased Bike"].value_counts()
plt.pie([purchased_bike["Yes"], purchased_bike["No"]])
```

```python
plt.xlabel('Posjeduju kucu')
plt.ylabel('Ne posjeduju kucu')
home_owners=bikes["Home Owner"].value_counts()
plt.pie([home_owners["Yes"], home_owners["No"]])
```
