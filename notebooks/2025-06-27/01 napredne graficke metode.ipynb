```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stanovi = pd.read_excel("stanovi_zagreb.xlsx")
```

```python
stanovi.head()
```

```python
stanovi.describe()
```

```python
stanovi.corr(numeric_only=True)
```

```python
plt.scatter(x=stanovi["povrsina"], y=stanovi["cijena"])
```

```python
sns.pairplot(stanovi)
```

```python
sns.boxplot(x = "cijena", y = "cetvrt", data = stanovi)
```

```python
cetvrti = stanovi["cetvrt"].value_counts()
```

```python
cetvrti.plot(kind="bar", color="blue")
```

```python
sns.barplot(data=cetvrti)
```

```python
cetvrti_kupaona = stanovi.groupby(['cetvrt', 'kupaona']).size().unstack(fill_value=0)
cetvrti_kupaona
```

```python
sns.barplot(data=cetvrti_kupaona, color="lightcoral")
```

```python
cetvrti_kupaona.plot(kind='bar', ax=plt.gca(), cmap='viridis')
```

```python
cetvrti_balkon = stanovi.groupby(['cetvrt', 'posjeduje balkon']).size().unstack(fill_value=0)
cetvrti_balkon
```

```python
sns.barplot(data=cetvrti_balkon, color="lightcoral")
```

```python
sns.barplot(data=cetvrti_balkon)
```

```python
cetvrti_balkon.plot(kind='bar', ax=plt.gca(), cmap='viridis')
```
